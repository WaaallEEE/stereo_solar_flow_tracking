```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Replace 'your_image.tiff' with the path to your TIFF file
light_f = '/Users/rattie/data/PUNCH/orion_green_channel.tif'
starless_f = '/Users/rattie/data/PUNCH/orion_green_channel_starless.tif'
# Open the image file
with Image.open(light_f) as img:
    # Convert the image to a NumPy array
    light = np.array(img)

with Image.open(starless_f) as img:
    # Convert the image to a NumPy array
    starless = np.array(img)

# Now 'image_array' is a NumPy array containing the image data
print(light.shape, light.mean(), light.min()) # This will print the dimensions of the array
print(starless.shape)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
axs[0].imshow(light, cmap='gray', vmin=1450, vmax=30000)
axs[1].imshow(starless, cmap='gray', vmin=1450, vmax=30000)
```
